# Use Feather format to share dataframes between R & pandas
- This is incredible! 
- https://blog.rstudio.com/2016/03/29/feather/

```python
import pandas as pd
import os
import feather
import sqlite3 as lite
```

```python
DATA_DIR = os.path.join(os.getcwd(), os.pardir, 'data')
feather_name = os.path.join(DATA_DIR, "iris.data")
iris_df = feather.read_dataframe(feather_name)
iris_df.head()
```

```python
iris_db_path = os.path.join(DATA_DIR, 'iris_db.sqlite')
conn = lite.connect(iris_db_path)
iris_df.to_sql('iris', conn, if_exists="replace", index=False)
conn.close()
```

```python
def execute_statement(query, database):
    ''' execute sql'''
    conn = lite.connect(database)
    result = pd.read_sql_query(query, conn)
    conn.close()
    return result
execute_statement("SELECT * FROM iris", iris_db_path).head(20)
```
